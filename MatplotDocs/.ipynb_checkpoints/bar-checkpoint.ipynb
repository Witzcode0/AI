# 📊 What is a Bar Chart?

A **bar chart** (or **bar graph**) is a graphical representation of data using **rectangular bars** to show the values of different categories.

---

## ✅ Key Points:

- Each **bar** represents a **category**.
- The **length or height** of the bar represents the **value** of that category.
- Bars can be **vertical** or **horizontal**.
- Useful for **comparing quantities** across different groups or categories.


```python
import matplotlib.pyplot as plt
```

```python

# ✅ Set a font that supports emojis (Windows: 'Segoe UI Emoji')
plt.rcParams['font.family'] = 'Segoe UI Emoji'

# ✅ Define categories with emojis
categories = ['🍎 Apples', '🍌 Bananas', '🍒 Cherries', '🌴 Dates', '🫐 Elderberries']

# ✅ Values corresponding to each fruit
values = [23, 45, 56, 78, 33]

# ✅ Custom colors for each bar
colors = ['blue', 'green', 'yellow', 'red', 'purple']

# ✅ Create a figure with a specific size
plt.figure(figsize=(10, 6))

# ✅ Create bar chart
bars = plt.bar(categories, values, color=colors, edgecolor='black', width=0.5)

# ✅ Legend is meant for plotted elements — remove or adjust below (see note)
plt.legend(['Average Line'], loc='upper left')  # Not needed here since no labeled plot added yet

# ✅ Add titles and axis labels
plt.title('Fruit Sales Report', fontsize=16)
plt.xlabel('Fruit Type', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)

# ✅ Add a light gray background to the chart area
plt.gca().set_facecolor('#f9f9f9')

# ✅ Add grid lines along Y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# ✅ Annotate each bar with its value above the bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, int(yval),
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# ✅ Customize the ticks
plt.xticks(rotation=45)
plt.yticks(range(0, 101, 10))

# ✅ Draw a horizontal line representing the average value
avg = sum(values) / len(values)
plt.axhline(avg, color='blue', linestyle='dashed', linewidth=1)

# ✅ Annotate the average line
plt.text(len(categories)-0.5, avg + 1, f'Avg: {avg:.1f}', color='blue')

# ✅ Automatically adjust layout to prevent overlap
plt.tight_layout()

# ✅ Display the plot
plt.show()
```
