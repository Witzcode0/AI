# What is a Bar Chart?

A **bar chart** (also called a **bar graph**) is a widely used data visualization tool that uses **rectangular bars** to represent values of different categories.  
The length or height of each bar corresponds to the magnitude of the data it represents.

---

## Key Characteristics of Bar Charts

- **Categorical Data:** Bar charts are ideal for visualizing data that can be grouped into categories.
- **Bars:** Each bar represents a **category**, and the **height or length** of the bar shows the corresponding value.
- **Orientation:** Bars can be displayed **vertically** or **horizontally**, depending on readability and label size.
- **Color/Pattern:** Colors, patterns, and textures can be used to differentiate bars or represent additional dimensions.
- **Simple to interpret:** Great for **quick comparisons** between categories.

---

## üìö Types of Bar Charts

| Type                        | Description                                                                 | Use Case Example                          |
|-----------------------------|-----------------------------------------------------------------------------|--------------------------------------------|
| **Vertical Bar Chart**      | Bars are displayed **upright**, X-axis = categories, Y-axis = values        | Sales per product                          |
| **Horizontal Bar Chart**    | Bars are displayed **horizontally**, Y-axis = categories, X-axis = values   | Responses with long labels or rankings     |
| **Grouped Bar Chart**       | Bars for multiple groups placed **side-by-side** within the same category   | Store A vs Store B per item                |
| **Stacked Bar Chart**       | Multiple bars **stacked** to show total + subgroup values                   | Revenue by product and region              |
| **100% Stacked Bar Chart**  | Like stacked bar, but each bar sums to **100%**                             | Market share composition                   |
| **Bar Chart with Error Bars** | Adds **error ranges** (e.g., ¬± standard deviation) to each bar             | Experimental or survey data                |
| **Patterned/Hatched Bars**  | Uses **patterns instead of colors** for fill                               | Black & white printing or accessibility    |
| **Sorted Bar Chart**        | Bars are sorted **ascending/descending** based on value                     | Top 10 products by sales                   |
| **Lollipop Chart**          | Minimal alternative to bar chart with a **dot and line**                   | Stylish or clean data presentation         |
| **Animated Bar Chart**      | Shows change over time via animation                                        | Race charts, time-series leaderboards      |

---

## üîç When to Use a Bar Chart?

Use a bar chart when you want to:

- Compare **quantities** across discrete categories
- Identify **trends**, **rankings**, or **outliers**
- Show **differences** between groups
- Visualize **part-to-whole** relationships (stacked bars)
- Present **experimental data** with uncertainty (error bars)

---

## üéØ Examples of Use Cases

- üõç Product sales by category  
- üßë‚Äçüè´ Test scores by subject  
- üåç Population by region  
- üß™ Experiment results with ¬± error  
- üìä Survey responses breakdown

---

## üß† Tip:

- Use **horizontal bars** when labels are too long for vertical orientation.
- Add **colors or emojis** for visual appeal.
- Use **average or threshold lines** (`axhline()` / `axvline()`) to enhance insights.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

# [1] Vertical Bar Chart with Emojis, Colors & Average Line

This chart visualizes fruit sales using a **vertical bar chart**, enhanced with:

- üé® Custom bar colors  
- üçá Emoji category labels  
- üìà Dashed average line  
- üßæ Value annotations above each bar  
- üí° Light background and grid for better readability

---

## ‚úÖ Key Features

- Uses `plt.bar()` for vertical bars
- Emojis improve category visualization
- `plt.axhline()` adds a horizontal average reference line
- `plt.text()` is used for bar value annotations
- Clean layout and tick formatting for professional look

---

```python

# ‚úÖ Set a font that supports emojis (Windows: 'Segoe UI Emoji')
plt.rcParams['font.family'] = 'Segoe UI Emoji'

# ‚úÖ Define categories with emojis
categories = ['üçé Apples', 'üçå Bananas', 'üçí Cherries', 'üå¥ Dates', 'ü´ê Elderberries']

# ‚úÖ Values corresponding to each fruit
values = [23, 45, 56, 78, 33]

# ‚úÖ Custom colors for each bar
colors = ['blue', 'green', 'yellow', 'red', 'purple']

# ‚úÖ Create a figure with a specific size
plt.figure(figsize=(10, 6))

# ‚úÖ Create bar chart
bars = plt.bar(categories, values, color=colors, edgecolor='black', width=0.5)

# ‚úÖ Legend is meant for plotted elements ‚Äî remove or adjust below (see note)
plt.legend(['Average Line'], loc='upper left')  # Not needed here since no labeled plot added yet

# ‚úÖ Add titles and axis labels
plt.title('Fruit Sales Report', fontsize=16)
plt.xlabel('Fruit Type', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)

# ‚úÖ Add a light gray background to the chart area
plt.gca().set_facecolor('#f9f9f9')

# ‚úÖ Add grid lines along Y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# ‚úÖ Annotate each bar with its value above the bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, int(yval),
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# ‚úÖ Customize the ticks
plt.xticks(rotation=45)
plt.yticks(range(0, 101, 10))

# ‚úÖ Draw a horizontal line representing the average value
avg = sum(values) / len(values)
plt.axhline(avg, color='blue', linestyle='dashed', linewidth=1)

# ‚úÖ Annotate the average line
plt.text(len(categories)-0.5, avg + 1, f'Avg: {avg:.1f}', color='blue')

# ‚úÖ Automatically adjust layout to prevent overlap
plt.tight_layout()

# ‚úÖ Display the plot
plt.show()
```

# [2] Horizontal Bar Chart with Emojis, Colors & Average Line

This chart displays a **horizontal bar graph** representing fruit sales.  
It includes:
- üé® Custom bar colors  
- üòÄ Emoji labels  
- üìà Average line  
- üßæ Value annotations on each bar  
- üåà Aesthetic background and grid  

---

## üîß Key Features
- `plt.barh()` is used for horizontal bars
- Emojis make category labels more fun and recognizable
- `plt.axvline()` is used to draw a vertical average line
- `plt.text()` is used to annotate bar values
- `plt.legend()` shows a label for the average line

---

```python
import matplotlib.pyplot as plt

# ‚úÖ Set a font that supports emojis (Windows: 'Segoe UI Emoji')
plt.rcParams['font.family'] = 'Segoe UI Emoji'

# ‚úÖ Define categories with emojis (Y-axis)
categories = ['üçé Apples', 'üçå Bananas', 'üçí Cherries', 'üå¥ Dates', 'ü´ê Elderberries']

# ‚úÖ Values corresponding to each fruit (X-axis values)
values = [23, 45, 56, 78, 33]

# ‚úÖ Custom colors for each bar
colors = ['blue', 'green', 'yellow', 'red', 'purple']

# ‚úÖ Create a figure with a specific size
plt.figure(figsize=(10, 6))

# ‚úÖ Create horizontal bar chart
bars = plt.barh(categories, values, color=colors, edgecolor='black')

# ‚úÖ Add titles and axis labels
plt.title('Fruit Sales Report (Horizontal)', fontsize=16)
plt.xlabel('Units Sold', fontsize=12)
plt.ylabel('Fruit Type', fontsize=12)

# ‚úÖ Add a light gray background to the chart area
plt.gca().set_facecolor('#f9f9f9')

# ‚úÖ Add grid lines along X-axis (because bars go horizontally)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# ‚úÖ Annotate each bar with its value at the end of the bar
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2.0, int(width),
             ha='left', va='center', fontsize=10, fontweight='bold')

# ‚úÖ Customize the ticks
plt.xticks(range(0, 101, 10))  # For X-axis values (units sold)
# No need to rotate Y-ticks for horizontal chart

# ‚úÖ Draw a vertical line representing the average value
avg = sum(values) / len(values)
plt.axvline(avg, color='blue', linestyle='dashed', linewidth=1, label='Average Line')

# ‚úÖ Annotate the average line
plt.text(avg + 1, -0.5, f'Avg: {avg:.1f}', color='blue', fontsize=10)

# ‚úÖ Add legend for average line
plt.legend(loc='lower right')

# ‚úÖ Automatically adjust layout to prevent overlap
plt.tight_layout()

# ‚úÖ Display the plot
plt.show()
```

# [3] Grouped Bar Chart (Side-by-Side)

A **grouped bar chart** shows multiple datasets side by side for each category.  
This makes it easy to **compare values across different groups** within the same category.

---

## ‚úÖ Why Use a Grouped Bar Chart?

- To compare **multiple groups** (e.g., Store A vs. Store B)
- See **differences within the same category**
- Useful for **sales comparison, survey results, or performance metrics**

---

```python
# Sample data
categories = ['Apples', 'Bananas', 'Cherries']
group1 = [30, 50, 70]  # e.g., Store A
group2 = [40, 60, 65]  # e.g., Store B

x = np.arange(len(categories))  # label locations
width = 0.35

plt.bar(x - width/2, group1, width, label='Store A')
plt.bar(x + width/2, group2, width, label='Store B')

plt.xticks(x, categories)
plt.title('Grouped Bar Chart')
plt.xlabel('Fruit')
plt.ylabel('Units Sold')
plt.legend()
```

# [4] Stacked Bar Chart in Matplotlib

A **stacked bar chart** displays multiple data series stacked on top of each other, in the same bar.  
This helps you visualize the **total value per category** and also the **contribution of each group** to that total.

---

## ‚úÖ Why Use a Stacked Bar Chart?

- To show **part-to-whole relationships** for each category
- Compare **multiple groups** across categories
- Easily see how each group contributes to the total

---

```python
group1 = [30, 50, 70]
group2 = [40, 60, 90]
group3 = [80, 120, 150]

plt.bar(categories, group1, label='Store A')
plt.bar(categories, group2, bottom=group1, label='Store B')
plt.bar(categories, group3, bottom=group2, label='Store C')

plt.title('Stacked Bar Chart')
plt.xlabel('Fruit')
plt.ylabel('Total Units Sold')
plt.legend()
```

# [5] 100% Stacked Bar Chart

A **100% stacked bar chart** is a type of bar chart where each bar represents **100% of a total**, and is split into parts to show the **relative percentage** of each sub-category.

Unlike regular stacked bar charts (which show actual values), **100% stacked charts normalize each bar** to total 100%, making it easier to compare **proportions** across categories.

---

## ‚úÖ Why Use a 100% Stacked Bar Chart?

- To visualize **composition** (part-to-whole relationship)
- To **compare proportions** across different groups, even when their total values vary
- Great for survey results, demographic breakdowns, etc.

---


```python
# Categories (e.g., products or regions)
categories = ['Product A', 'Product B', 'Product C']

# Raw data (e.g., sales by region or response count)
group1 = np.array([30, 50, 20])   # Region 1
group2 = np.array([20, 30, 50])   # Region 2
group3 = np.array([50, 20, 30])   # Region 3

# Sum per category
totals = group1 + group2 + group3

# Convert to percentage (each bar adds up to 100)
group1_percent = group1 / totals * 100
group2_percent = group2 / totals * 100
group3_percent = group3 / totals * 100

# X-axis positions
x = np.arange(len(categories))

# Plotting
plt.figure(figsize=(10, 6))

plt.bar(x, group1_percent, label='Region 1')
plt.bar(x, group2_percent, bottom=group1_percent, label='Region 2')
plt.bar(x, group3_percent, bottom=group1_percent + group2_percent, label='Region 3')

# Labels and formatting
plt.title('üíØ 100% Stacked Bar Chart')
plt.ylabel('Percentage (%)')
plt.xticks(x, categories)
plt.yticks(np.arange(0, 101, 10))
plt.legend(loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display percentages on each section (optional)
for i in range(len(categories)):
    plt.text(x[i], group1_percent[i]/2, f'{group1_percent[i]:.0f}%', ha='center', va='center', color='white', fontsize=8)
    plt.text(x[i], group1_percent[i] + group2_percent[i]/2, f'{group2_percent[i]:.0f}%', ha='center', va='center', color='white', fontsize=8)
    plt.text(x[i], 100 - group3_percent[i]/2, f'{group3_percent[i]:.0f}%', ha='center', va='center', color='white', fontsize=8)

plt.tight_layout()
plt.show()
```

# [6] Bar Chart with Error Bars

A **bar chart with error bars** is used to show not only the values for each category, but also the **uncertainty or variation** in those values.

---

## ‚úÖ What are Error Bars?

- **Error bars** represent the **range of possible values** (e.g., ¬± standard deviation or margin of error).
- They help visualize how **precise or reliable** each data point is.
- In `matplotlib`, you can add error bars using the `yerr` parameter in `plt.bar()`.

---

## üçè Example: Error Ranges Per Category

Each bar will have a vertical line above and/or below it, like:

- üçé Apples ‚Üí 23 ¬± 2  
- üçå Bananas ‚Üí 45 ¬± 5  
- üçí Cherries ‚Üí 56 ¬± 3  
- üå¥ Dates ‚Üí 78 ¬± 4  
- ü´ê Elderberries ‚Üí 33 ¬± 1  

This means:
- **Apples** might actually be between **21 and 25**
- **Bananas** might be between **40 and 50**, and so on.

---

```python
errors = [2, 5, 3, 4, 1]
# ‚úÖ Define categories with emojis (Y-axis)
categories = ['üçé Apples', 'üçå Bananas', 'üçí Cherries', 'üå¥ Dates', 'ü´ê Elderberries']

# ‚úÖ Values corresponding to each fruit (X-axis values)
values = [23, 45, 56, 78, 33]
plt.bar(categories, values, yerr=errors, capsize=5, color='skyblue', edgecolor='black')
plt.title("Bar Chart with Error Bars")
plt.show()
```

# [7] Bar Chart with Patterns (Hatching)

**Hatching (patterns)** in bar charts is used to fill bars with textures like slashes (`/`), backslashes (`\`), vertical lines (`|`), dashes (`-`), or plus signs (`+`).

---

## ‚úÖ Why Use Patterns?

- Useful when printing in **black and white**
- Improves **readability** for people with **color vision deficiency**
- Adds **style** and differentiation beyond color

---

## üîç Common Hatch Patterns

| Pattern | Symbol | Description            |
|---------|--------|------------------------|
| `/`     | Slash  | Diagonal lines (right) |
| `\`     | Backslash | Diagonal lines (left) |
| `|`     | Pipe   | Vertical lines         |
| `-`     | Dash   | Horizontal lines       |
| `+`     | Plus   | Crosshatch             |

---


```python
# Categories and values
categories = ['üçé Apples', 'üçå Bananas', 'üçí Cherries', 'üå¥ Dates', 'ü´ê Elderberries']
values = [23, 45, 56, 78, 33]

# Hatch patterns for each bar
patterns = ['/', '\\', '|', '-', '+']

# Create bar chart with patterns
plt.figure(figsize=(10, 6))

for i in range(len(categories)):
    plt.bar(categories[i], values[i],
            hatch=patterns[i],
            color='white',
            edgecolor='black',
            linewidth=1.2)

# Add titles and labels
plt.title("Bar Chart with Hatch Patterns", fontsize=16)
plt.xlabel("Fruit Type")
plt.ylabel("Units Sold")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

# [8] Sorted Bar Chart

A **sorted bar chart** arranges bars in **ascending or descending order** based on their values.  
This helps highlight **rankings** or emphasize **extremes** like highest or lowest values.

### ‚úÖ Use Cases:
- Top-selling products
- Highest scoring students
- Ranking countries by population

```python
# Sample data
categories = ['Bananas', 'Apples', 'Dates', 'Cherries', 'Elderberries']
values = [45, 23, 78, 56, 33]

# Sort data in descending order
sorted_data = sorted(zip(values, categories), reverse=True)
sorted_values, sorted_categories = zip(*sorted_data)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_categories, sorted_values, color='skyblue', edgecolor='black')
plt.title('üî¢ Sorted Bar Chart (Descending)', fontsize=14)
plt.ylabel('Quantity Sold')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# [9] Lollipop Chart

A **Lollipop Chart** is a modern variation of a bar chart that uses a **line and a dot** instead of a full rectangular bar.  
It communicates the same information but with a **cleaner, minimalistic look**, making it a great choice for dashboards and presentations.

---

## ‚úÖ Why Use a Lollipop Chart?

- Reduces visual clutter by replacing bars with thin lines
- Makes charts look **cleaner and more modern**
- Ideal for **print materials** or minimalist design
- Easier to highlight exact values with dots

---

## üìå Use Cases

- ‚úÖ **Minimalistic dashboards**  
- ‚úÖ **Rankings with less ink**  
- ‚úÖ **Alternative to bar charts for clean reports or presentations**  
- ‚úÖ **Top-N comparisons** (e.g., top 10 cities, highest scoring students)

---

```python
# Sample data
categories = ['üçé Apples', 'üçå Bananas', 'üçí Cherries', 'üå¥ Dates', 'ü´ê Elderberries']
values = [23, 45, 56, 78, 33]
positions = range(len(categories))

# Create plot
plt.figure(figsize=(10, 6))
plt.hlines(y=positions, xmin=0, xmax=values, color='skyblue')     # horizontal lines
plt.plot(values, positions, 'o', color='darkblue')                 # lollipop heads (dots)

# Label and format
plt.yticks(positions, categories)
plt.xlabel('Units Sold')
plt.title('üéØ Lollipop Chart: Fruit Sales')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# [10] Animated Bar Chart

An **Animated Bar Chart** is a dynamic data visualization that shows how values **change over time**.  
Each frame in the animation represents a **time step** (e.g., day, month, year), making it ideal for showing **trends, progress, or race charts**.

---

## ‚úÖ Why Use Animated Bar Charts?

- Helps **visualize progression** or **competition** over time
- Makes data **more engaging** and easier to interpret
- Useful for **storytelling**, dashboards, or social media visuals

---

## üìå Use Cases

- üìà **Sales** growth per month or quarter  
- üë• **Social media followers** over time  
- üèÅ **Race charts** showing top competitors dynamically  
- üíπ **Stock or revenue** trends by year  
- ‚öôÔ∏è **System/resource usage** logs visualization

---


```python
from IPython.display import HTML

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [20, 35, 50, 70, 90]

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(['Sales'], [0], color='orange')  # Start with 0 height
ax.set_ylim(0, max(sales) + 10)
ax.set_title("üïò Animated Sales Chart")
ax.set_ylabel("Units Sold")

# Update function for each frame
def update(frame):
    bars[0].set_height(sales[frame])
    ax.set_xlabel(f"Month: {months[frame]}")

# Create animation
anim = animation.FuncAnimation(fig, update, frames=len(sales), interval=1000, repeat=False)

# Render it safely using to_jshtml()
HTML(anim.to_jshtml())
```
