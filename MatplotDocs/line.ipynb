# What is a Line Chart?

A **line chart** (or **line graph**) is a basic chart used to display **trends over time** or across ordered data points by connecting individual data values with straight lines.

---

## ‚úÖ Key Features

- Each **point** represents a data value.
- Points are connected by **lines** to show movement/trend.
- Ideal for **time-series** or **continuous data**.

---

## üìå Use Cases

- Tracking **sales over months**
- Monitoring **temperature over time**
- Visualizing **stock price trends**
- Showing **growth patterns** (users, revenue, etc.)

---

## üß† Advantages

- Clear **visual of trends**
- Can compare **multiple series**
- Works well with **dense data**

---

## üìä Common Types of Line Charts

| Type                   | Description                                               |
|------------------------|-----------------------------------------------------------|
| **Basic Line Chart**   | Single series connected by a line                         |
| **Multi-Line Chart**   | Multiple series shown together for comparison             |
| **Line with Markers**  | Adds dots or shapes at each data point                    |
| **Smoothed Line**      | Uses curves (e.g., spline) to smooth the line             |
| **Filled Line Area**   | Area under the line is filled with color (Area Chart)     |
| **Step Line**          | Data changes in steps rather than slopes                  |
| **Dashed/Styled Line** | Custom line styles (dashed, dotted) for better comparison |

---

```python
# !pip install numpy
# !pip install pandas
# !pip install matplotlib
# !pip install scipy
# !pip install plotly

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

## 1Ô∏è‚É£ Basic Line Chart
Shows a single series of data connected by lines.

**Use Case:** Simple trend over time (e.g., monthly sales)

---

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [20, 35, 50, 45, 60]

plt.figure(figsize=(8, 4))
plt.plot(months, sales, color='blue')
plt.title('Basic Line Chart')
plt.grid(True)
plt.show()
```

## 2Ô∏è‚É£ Line Chart with Markers
Adds markers to each data point to highlight exact values.

**Use Case:** Spotting exact values at key time intervals

---

```python
plt.figure(figsize=(8, 4))
plt.plot(months, sales, color='green', marker='o')
plt.title('Line Chart with Markers')
plt.grid(True)
plt.show()
```

## 3Ô∏è‚É£ Multi-Line Chart
Compares multiple datasets on the same chart.

**Use Case:** Compare sales of two stores month-wise

---

```python
store_a = [20, 35, 50, 45, 60]
store_b = [25, 40, 55, 50, 70]

plt.figure(figsize=(8, 4))
plt.plot(months, store_a, label='Store A', marker='o')
plt.plot(months, store_b, label='Store B', marker='s')
plt.title('Multi-Line Chart')
plt.legend()
plt.grid(True)
plt.show()
```

## 4Ô∏è‚É£ Smoothed Line Chart (Curved)
Uses interpolation to make curves instead of straight lines.

**Use Case:** More aesthetically pleasing trend lines

---

```python
from scipy.interpolate import make_interp_spline

x = np.arange(len(months))
y = np.array(sales)
x_smooth = np.linspace(x.min(), x.max(), 300)
spline = make_interp_spline(x, y, k=3)
y_smooth = spline(x_smooth)

plt.figure(figsize=(8, 4))
plt.plot(x_smooth, y_smooth, color='darkorange')
plt.xticks(x, months)
plt.title('Smoothed Line Chart')
plt.grid(True)
plt.show()
```

## 5Ô∏è‚É£ Area Line Chart (Filled)
Fills the area under the line to show volume/intensity.

**Use Case:** Show sales volume visually

---


```python
plt.figure(figsize=(8, 4))
plt.fill_between(months, sales, color='skyblue', alpha=0.5)
plt.plot(months, sales, color='blue')
plt.title('Area Line Chart')
plt.grid(True)
plt.show()
```

## 6Ô∏è‚É£ Step Line Chart
Draws a step between data points (not diagonal).

**Use Case:** Stock price changes at discrete intervals

---

```python
plt.figure(figsize=(8, 4))
plt.step(months, sales, where='mid', color='purple', marker='o')
plt.title('Step Line Chart')
plt.grid(True)
plt.show()
```

## 7Ô∏è‚É£ Dashed/Styled Line Chart
Applies different line styles (dotted, dashed, etc.).

**Use Case:** Distinguish categories visually

---

```python
plt.figure(figsize=(8, 4))
plt.plot(months, sales, linestyle='--', color='brown', marker='d')
plt.title('Dashed Line Style')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
import plotly.io as pio

# Set renderer for your environment
pio.renderers.default = 'notebook'  # try 'iframe' or 'browser' if 'notebook' doesn't work

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [20, 35, 50, 45, 60]

fig = px.line(x=months, y=sales, markers=True, title='üìà Interactive Line Chart')
fig.update_layout(xaxis_title='Month', yaxis_title='Sales')
fig.show()
```

# üîç Multi-Series Line Plot with Styling and Fill

This example demonstrates a detailed multi-line chart with:

- ‚ûï Red plus markers for one series
- üî∫ Green triangle markers for another
- üîµ Blue dashed line with circular markers
- üíß Area filled between one line and the x-axis
- ‚úÖ Axis labels, limits, title, and grid for better visualization

---

## üìä Features Used:

- `plt.plot(x, y, 'r+')`: Red plus (`+`) markers
- `plt.plot(x, y3, 'g^')`: Green triangle-up markers
- Styled plot for `y2`: Dashed blue line with large magenta-filled circles
- `plt.fill_between(x, y, ...)`: Fills area under `y` values
- Axis range controlled via `plt.axis([xmin, xmax, ymin, ymax])`
- Grid enabled using `plt.grid()`


```python
# Sample X values
x = [1, 5, 8, 9]

# Three sets of Y values for three different plots
y = [55, 22, 44, 88]   # Red '+' markers
y2 = [60, 27, 49, 93]  # Blue solid line
y3 = [65, 32, 54, 98]  # Green triangle-up markers

# Plot 1: x vs y
# - 'r+' ‚Üí red color, plus (+) markers
plt.plot(x, y, "r+",    # Red '+' markers
         x, y3, "g^")   # Green triangle-up (^) markers

# Plot 2: x vs y2 with detailed styling
plt.plot(
    x, y2, 
    color='blue',          # Line color: blue
    linestyle='dashed',    # Line style: dashed
    marker='o',            # Marker style: circle
    markerfacecolor='magenta',  # Fill color for markers
    markersize=12,         # Marker size
    linewidth=2,            # Line width
)

# Title of the plot
plt.title("Title")

# Set the limit for X-axis from 0 to 10
plt.xlim([0, 10])

# Set the limit for Y-axis from 0 to 100
plt.ylim([0, 100])

# Alternatively, you can use plt.axis() to set all four limits at once
# Format: plt.axis([xmin, xmax, ymin, ymax])
plt.axis([0, 10, 0, 100])

# Note: If you use both plt.xlim()/plt.ylim() and plt.axis(), the last one called will take effect.

# Adding labels to axes
plt.xlabel("X-Label")
plt.ylabel("Y-Label")

# Fill the area between the line (y) and the x-axis
# - color='skyblue': fill color
# - alpha=0.4: transparency level (0 = fully transparent, 1 = fully opaque)
plt.fill_between(x, y, color='skyblue', alpha=0.4)

# Show grid for better readability
plt.grid()

# Display the plot
plt.show()
```

# üîç Subplot Example: Natural Logarithm and Its Square

This example demonstrates how to use `plt.subplot()` to create two plots side-by-side:

---

## üìä Plot 1: `y = log(x)`

- Shows the natural logarithm of values from 1 to 10
- Smoothly increases but slows down as x increases

---

## üìä Plot 2: `y = log(x)^2`

- Square of the logarithm
- Useful in transformations and statistical visualizations

---

- Both subplots are placed using:
  - `plt.subplot(121)` ‚Üí 1 row, 2 columns, plot 1
  - `plt.subplot(122)` ‚Üí 1 row, 2 columns, plot 2

- Grid and axis labels are added for clarity.


```python
# üîπ Create 100 evenly spaced values from 1 to 10
x = np.linspace(1, 10, 100)

# üîπ Compute the natural logarithm of x
y = np.log(x)

# üîπ First subplot: y = log(x)
plt.subplot(121)  # (1 row, 2 columns, position 1)
plt.plot(x, y, color='blue')
plt.title("y = log(x)")
plt.xlabel("x")
plt.ylabel("log(x)")
plt.grid(True)

# üîπ Second subplot: y = log(x)^2
plt.subplot(122)  # (1 row, 2 columns, position 2)
plt.plot(x, y**2, color='green')
plt.title("y = log(x)^2")
plt.xlabel("x")
plt.ylabel("log(x)^2")
plt.grid(True)

# üîπ Show both plots side-by-side
plt.tight_layout()
plt.show()
```

# üìê Comparing Polynomial Growth

This example demonstrates how values of `x` grow at different polynomial degrees:

| Function      | Plot Color | Figure | Subplot |
|---------------|------------|--------|---------|
| `y = x`       | Green      | 1      | Top-Left |
| `y = x¬≤`      | Blue       | 1      | Top-Right |
| `y = x¬≥`      | Orange     | 2      | Top-Left |
| `y = x‚Å¥`      | Cyan       | 2      | Top-Right |

---

## ‚úÖ Concepts Covered

- **`lambda` with `map()`** to compute powers
- **Multiple figures** using `plt.figure()`
- **Subplots** layout via `plt.subplot(rows, cols, pos)`
- **Legends and gridlines** for clarity
- **Tight layout** to avoid label overlap


```python
# X values
x = [1, 2, 3, 4, 5]

# Compute powers of x using lambda and map
xp2 = list(map(lambda x: x**2, x))  # x^2
xp3 = list(map(lambda x: x**3, x))  # x^3
xp4 = list(map(lambda x: x**4, x))  # x^4

# üìä Figure 1: Linear and Square
plt.figure(1)

plt.subplot(2, 2, 1)  # Top-left
plt.title("y = x (Linear)")
plt.plot(x, x, color="green", label="Linear")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)  # Top-right
plt.title("y = x¬≤ (Square)")
plt.plot(x, xp2, color="blue", label="Squared")
plt.grid(True)
plt.legend()

plt.tight_layout()  # Fix spacing
plt.figure(2)

# üìä Figure 2: Cube and Fourth Power
plt.subplot(2, 2, 1)  # Top-left
plt.title("y = x¬≥ (Cube)")
plt.plot(x, xp3, color="orange", label="Cubed")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)  # Top-right
plt.title("y = x‚Å¥ (Fourth Power)")
plt.plot(x, xp4, color="cyan", label="Fourth Power")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# üìÖ Time Series Line Plot using Matplotlib

This plot visualizes a simple time series over five consecutive dates starting from **January 1, 2023**.

---

## ‚úÖ Key Features:

- Dates generated using `pd.date_range()`
- Line plot with **circular markers**
- Axis labels, title, and rotated date ticks
- Gridlines for readability
- Saved as a high-resolution image using `uuid` for a unique filename

---

## üß™ Use Case:

Time series plots are useful for:
- Stock price trends
- Sensor readings over time
- Daily/weekly analytics


```python
import uuid
import matplotlib.dates as mdates

# Create 5 consecutive dates starting from Jan 1, 2023
dates = pd.date_range("2023-01-01", periods=5)

# Corresponding values for each date
values = [10, 20, 15, 25, 30]

# Create the plot
plt.figure(figsize=(8, 5))            # Optional: Set figure size
plt.plot(dates, values, marker='o', linestyle='-', color='teal', label="Daily Values")

# Add title and axis labels
plt.title("Time Series Plot", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Value")

# Rotate date labels for better readability
plt.xticks(rotation=30)

# Format x-axis ticks as "Jan 01" format
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

# Show grid
plt.grid(True)

# Show legend (optional since only one line)
plt.legend()

# Save the plot as a high-resolution PNG file
plt.savefig(f"{uuid.uuid4()}.png", dpi=300)

# Show the plot
plt.show()
```
