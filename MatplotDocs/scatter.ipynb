# ðŸŽ¯ What is a Scatter Plot?

A **scatter plot** is used to display values for two variables for a set of data. Each point represents a single observation with its x and y values.

---

## âœ… Key Features:

- Each **dot** represents one data point.
- X-axis and Y-axis represent two **numerical variables**.
- Useful for **finding correlations**, **clusters**, or **outliers**.
- Can be enhanced using **colors**, **sizes**, or **shapes**.

---

## ðŸ“Œ Common Use Cases:

- Comparing height vs. weight
- Examining exam score vs. study time
- Clustering in ML datasets
- Visualizing relationships in data

---

## ðŸ“Š Types of Scatter Plots:

| Type                      | Description                                         |
|---------------------------|-----------------------------------------------------|
| Basic Scatter Plot        | Simple dots for (x, y) pairs                        |
| Colored Scatter Plot      | Different colors based on category/class            |
| Size-Scaled Scatter Plot  | Dot size varies with a third variable               |
| Scatter Matrix            | Grid of scatter plots for multiple variables        |
| 3D Scatter Plot           | Uses a third axis (Z) for additional information    |
| Bubble Plot               | Like scatter plot, but with size and color together |



# Scatter Plot Types :

Scatter plots visualize the relationship between two or more numeric variables using dots. Useful for identifying patterns, clusters, trends, and outliers.

---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Basic Scatter Plot

**ðŸ”¹ Description:**  
Displays individual data points using dots based on two variables (x, y).

**ðŸ“Œ Use Case:**  
Compare study hours vs marks, height vs weight, etc.

---

```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

## 2. âœ… Colored Scatter Plot

**ðŸ”¹ Description:**  
Each dot is assigned a specific color to represent a category or class.

**ðŸ“Œ Use Case:**  
Group data points by gender, region, product type, etc.

---


```python
colors = ['red', 'green', 'blue', 'purple', 'orange']

plt.scatter(x, y, c=colors)
plt.title("Colored Scatter Plot")
plt.show()
```

## 3. âœ… Size-Scaled Scatter Plot

**ðŸ”¹ Description:**  
Point size varies based on a third numeric variable.

**ðŸ“Œ Use Case:**  
Show population, revenue, or quantity using dot size.

---

```python
sizes = [20, 50, 80, 200, 500]

plt.scatter(x, y, s=sizes)
plt.title("Size-Scaled Scatter Plot")
plt.show()
```

## 4. âœ… Bubble Plot (Color + Size)

**ðŸ”¹ Description:**  
Uses both color and size of points to encode 4-dimensional data.

**ðŸ“Œ Use Case:**  
Market analysis by sales (size), profit (position), region (color).

---

```python
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, edgecolors='black')
plt.title("Bubble Plot")
plt.show()
```

## 5. âœ… Scatter Matrix Plot

**ðŸ”¹ Description:**  
A grid of scatter plots comparing all pairs of features in a multivariate dataset.

**ðŸ“Œ Use Case:**  
Explore relationships between all numerical columns (EDA).

---

```python
from pandas.plotting import scatter_matrix

data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

scatter_matrix(df, figsize=(8, 6), diagonal='hist')
plt.suptitle("Scatter Matrix")
plt.show()
```

## 6. âœ… 3D Scatter Plot

**ðŸ”¹ Description:**  
Adds a Z-axis to display relationships among three continuous variables.

**ðŸ“Œ Use Case:**  
Visualize GDP, literacy rate, and life expectancy in one chart.

---

```python
from mpl_toolkits.mplot3d import Axes3D

z = [5, 4, 3, 2, 1]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

ax.set_title("3D Scatter Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```

## 7. âœ… Animated Scatter Plot

**ðŸ”¹ Description:**  
Dynamic plot showing how data points evolve over time.

**ðŸ“Œ Use Case:**  
Track monthly sales, live user growth, or time-series competition.

---

```python
import matplotlib.animation as animation
from IPython.display import HTML

fig, ax = plt.subplots()
y_vals = [[2, 4, 1, 3, 5], [3, 5, 2, 4, 6], [4, 6, 3, 5, 7]]
sc = ax.scatter(x, y_vals[0])
ax.set_xlim(0, 6)
ax.set_ylim(0, 8)

def update(i):
    sc.set_offsets(list(zip(x, y_vals[i])))

ani = animation.FuncAnimation(fig, update, frames=len(y_vals), interval=1000, repeat=False)
HTML(ani.to_jshtml())
```
